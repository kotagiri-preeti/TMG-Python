```python
#create a file in write mode
f = open("demo.txt","w")
```

```python
#write mode --> Removes previous content and add only new content
f1 = open("demo.txt","w")
f1.write("Good Evening!!!!")
f1.close()
```

```python
#append mode --> Previous content remains as it is 
#and new content is added at the end
f2 = open("demo.txt","a")
f2.write("Hello Everyone....How are you doing all???")
f2.close()
```

```python
#Reading a file
f3 = open("demo.txt","r")
data = f3.read()
print(data)
print(type(data))
f3.close()
```
